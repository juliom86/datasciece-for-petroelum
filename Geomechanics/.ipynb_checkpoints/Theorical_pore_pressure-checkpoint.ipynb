# Ejercicio teorico de presion poral


## Importo librerias


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Genero la data

```python
d = np.arange(20000,-1000,-1000)

rhob = np.array([2.48,2.47,2.46,2.45,2.44,2.43,2.41,2.40,2.38,2.37,2.35,2.33,2.29,2.27,2.24,2.19,2.16,2.11,2.06,2.02,1.95])

rhog = 2.65 #g/cm3

rhof = 1.1 

phi = (rhog - rhob)/(rhog - rhof)
```

Una vez generado, procedo a graficar

```python
plt.style.use('default')
plt.figure(figsize=(3,6))

plt.plot(rhob,d,label='RHOB')
plt.plot(phi,d,label='PHI')

plt.ylim(20000,0)
# plt.xlabel('bulk density')
plt.ylabel('Profundidad')
plt.legend()

plt.title('Determinacion de la Porosidad')
plt.grid()
```

```python
y = d1[d1<=12500]
x = dc[0:15]

# z = np.polyfit(x, y, 1)
# p = np.poly1d(z)
d11 = np.array([6000,8100,9000,9600,10100,10400,10700,10900,11100,11300,11500,11600,11800,12100,12200,12300,
               12700,12900,13000,13200,13400,13500,13600,13700,13800,13900,14000,
               14200,14400,14600,14800,14900,15000,15200,15300,15400,15500,15700,16200,16800])
dmod = 1.15 + 0.000038*d11  #dmod = 1.15 + 0.000038D
```

## Graficamos las Curvas teoricas y generadas artificialmente

```python
plt.style.use('default')
plt.figure(figsize=(3,6))

plt.scatter(dc,d1,label='d-exp(normal)')
plt.scatter(x,y,label='d(normal)-linea de tendencia')
plt.plot(dmod,d11,'red',label='d-exp(corregida)')
plt.ylim(16800,5000)

plt.legend(loc='upper left')
plt.title('Presion poral calculada por metodo exponencial')
plt.ylabel('Depth(ft)')
plt.grid()
```
